```python
import tabula
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'table_of_judges_by_year.pdf'
```

```python
df = tabula.read_pdf(file, pages='all')
```

```python
df.head()
```

```python
df = df[['Unnamed: 0', 'Unnamed: 4']]
df.columns = ['Year','Auth. Dist. Ct. Judgeships']
df.drop(labels=[0,1,2,3], axis=0, inplace=True)
df.set_index('Year', inplace=True,drop=True)
df.head(10)
```

```python
df.to_csv('authorized_judgeships_by_year.csv')
```
